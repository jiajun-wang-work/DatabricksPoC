```python
from pyspark.sql import SparkSession

# SparkSessionの作成
spark = SparkSession.builder \
    .appName("example_app") \
    .getOrCreate()

# DataFrameの作成
data = [("Alice", 34), ("Bob", 45), ("Charlie", 25)]
df = spark.createDataFrame(data, ["Name", "Age"])

# DataFrameの表示
df.show()

# スキーマの表示
df.printSchema()

# データのフィルタリング
filtered_df = df.filter(df.Age > 30)
filtered_df.show()

# DataFrameを一時的なビューとして登録
df.createOrReplaceTempView("people")

# SQLクエリの実行
result = spark.sql("SELECT * FROM people WHERE Age > 30")

# 結果の表示
result.show()
```

```python
# データの集計
from pyspark.sql.functions import avg
average_age = df.select(avg(df.Age)).collect()[0][0]
print("Average Age:", average_age)

# 新しい列の追加
from pyspark.sql.functions import lit
df = df.withColumn("Department", lit("IT"))
df.show()
```

```python
from hdfs import InsecureClient

# HDFSクライアントの作成
client = InsecureClient("http://localhost:50070", user="hadoop")

# サンプルデータの作成
data = [("Alice", 34), ("Bob", 45), ("Charlie", 25)]
data_str = "\n".join([f"{name},{age}" for name, age in data])

# データをHDFSのData Lakeに保存
with client.write("/path/to/data_lake/sample_data.txt", overwrite=True) as writer:
    writer.write(data_str)

print("Data saved to Data Lake.")
```

```python
from pyspark.sql import SparkSession

# SparkSessionの作成
spark = SparkSession.builder \
    .appName("Delta Lake Example") \
    .getOrCreate()

# サンプルデータの作成
data = [("Alice", 34), ("Bob", 45), ("Charlie", 25)]
df = spark.createDataFrame(data, ["Name", "Age"])

# データをDelta Lakeテーブルに書き込む
df.write.format("delta").save("/path/to/delta_table")

# Delta Lakeテーブルからデータを読み込む
df_read = spark.read.format("delta").load("/path/to/delta_table")

# データの表示
df_read.show()
```

```python
from pyspark.sql import SparkSession

# SparkSessionの作成
spark = SparkSession.builder \
    .appName("DataFrameReaderWriterExample") \
    .getOrCreate()

# CSVファイルを読み込んでDataFrameに変換
df = spark.read.csv("/path/to/input.csv", header=True, inferSchema=True)

# DataFrameをParquetファイルとして書き込む
df.write.parquet("/path/to/output.parquet")

# ParquetファイルからDataFrameを読み込む
df_read = spark.read.parquet("/path/to/output.parquet")

# データの表示
df_read.show()

# SparkSessionの終了
spark.stop()
```

```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("DropNAExample").getOrCreate()

# Example list of tuples with null values
data = [(1, 'Alice', None), (2, None, 'Jones'), (3, 'Charlie', 'Brown')]

# Create DataFrame from list of tuples
df = spark.createDataFrame(data, ['id', 'first_name', 'last_name'])

# Drop rows with any null values
df_no_nulls = df.dropna()

# Show the resulting DataFrame
df_no_nulls.show()
```

```python
# Fill null values with a specific value
df_filled = df.fillna({'first_name': 'Unknown', 'last_name': 'Unknown'})

# Show the resulting DataFrame
df_filled.show()
```

```python
import time
from functools import wraps

def timing_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Function {func.__name__} took {end_time - start_time:.4f} seconds to execute.")
        return result
    return wrapper

@timing_decorator
def some_pyspark_operation(data):
    df = spark.createDataFrame(data, ['id', 'first_name', 'last_name'])
    df_no_duplicates = df.dropDuplicates()
    return df_no_duplicates

# Example list of tuples
data = [(1, 'Alice'), (2, 'Bob'), (1, 'Alice'), (3, 'Charlie')]

# Call the function with the decorator applied
result_df = some_pyspark_operation(data)
result_df.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import regexp_replace

# Initialize Spark session
spark = SparkSession.builder.appName("ReplaceSpaceWithHyphen").getOrCreate()

# Example list of tuples
data = [(1, 'Alice Smith'), (2, 'Bob Jones'), (3, 'Charlie Brown')]

# Create DataFrame from list of tuples
df = spark.createDataFrame(data, ['id', 'name'])

# Replace spaces with hyphens in the 'name' column
df_with_hyphens = df.withColumn('name', regexp_replace(df.name, ' ', '-'))

# Show the resulting DataFrame
df_with_hyphens.show()
```

```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("JoinExample").getOrCreate()

# Example data
data1 = [(1, 'Alice'), (2, 'Bob')]
data2 = [(1, 'HR'), (2, 'Finance'), (3, 'IT')]

# Create DataFrames
df1 = spark.createDataFrame(data1, ['id', 'name'])
df2 = spark.createDataFrame(data2, ['id', 'department'])

# Perform an inner join
df_joined = df1.join(df2, on='id', how='inner')

# Show the result
df_joined.show()
```

```python
# Example data
data3 = [(3, 'Charlie'), (4, 'David')]

# Create another DataFrame
df3 = spark.createDataFrame(data3, ['id', 'name'])

# Concatenate DataFrames
df_concatenated = df1.union(df3)

# Show the result
df_concatenated.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import array, lit

# Initialize Spark session
spark = SparkSession.builder.appName("ArrayExample").getOrCreate()

# Example data
data = [(1, "Alice", 10, 20), (2, "Bob", 30, 40), (3, "Charlie", 50, 60)]

# Create DataFrame
df = spark.createDataFrame(data, ["id", "name", "score1", "score2"])

# Create an array column from score1 and score2
df_with_array = df.withColumn("scores", array("score1", "score2"))

# Show the resulting DataFrame
df_with_array.show()
```
